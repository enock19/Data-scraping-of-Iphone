```python
import requests
from bs4 import BeautifulSoup
```

```python
renponse = requests.get("https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=iphone&_sacat=0")
print(renponse)
```

```python
soup=BeautifulSoup(renponse.text,"html.parser")
print(soup)
```

```python
title=soup.find_all("div",class_="s-item__title")
products=[]
for title in title:
    products.append(title.text)
del products[0]
print(products)
len(products)
```

```python
price=soup.find_all("span",class_="s-item__price")
description=[]
for price in price:
    description.append(price.text)
del description[0]
print(description)
len(description)
```

```python
import pandas as pd
```

```python
new_data=pd.DataFrame({"title":products,"price":description})
```
